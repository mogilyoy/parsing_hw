### dzen.ru

```python
url = 'https://dzen.ru/'
user_agent = 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/109.0'

headers={
    'User-Agent': user_agent,
}

response = requests.get(url, headers=headers, allow_redirects=True)
```

```python
start = response.text.find('uuid=')
finish = response.text.find('","retpath":')

uuid = response.text[start+5:finish]
response2 = requests.get(f'https://sso.dzen.ru/install?uuid={uuid}', headers=headers)
```

```python
dom = html.fromstring(response2.text)
```

```python
titles = dom.xpath("//div[@class='card-news__title-2w']/text()")
titles
```

```python
a = dom.xpath("//a[@class='news-story__link-14']")
```

```python
desc = []

for i in a:
    desc.append(i.xpath("./div[@class='news-story__textContainer-Do']/span/text()")[0])
desc[10]
```

```python
hrefs = []
for i in a:
    hrefs.append(i.xpath("@href")[0])

hrefs[15]
```

```python
result=[]


for i in range(len(titles)):
    for j in range(i*5, (i+1)*5):
        result.append({
            'id': j,
            'group': titles[i],
            'url': hrefs[j],
            'title': desc[j],
        })

result[0:3]
```
